## Câu lệnh if..else

```python
mark = float(input("nhap so diem cuoi ky: "))
if mark >= 8.5:
    print("ban duoc tuyen thang")
else:
    print("ban chua du diem")
    
```

### Thực hành:
Viết chương trình nhập vào một số nguyên và cho biết số nguyên đó là số chẵn hay số lẻ.

```python
number = int(input("nhap vao mot so nguyen "))
if number % 2 == 0:
    print(f"{number} la mot so chan")
else:
    print(f"{number} la so le")
```

### Thực hành:
Viết chương trình nhập vào một chuỗi và cho biết trong chuỗi đó có chứa cụm từ "python" hay không.

```python
string = str(input("please input a string :"))
if string.count("python") > 0:
    print(f'"python" exists in {string}')
else:
    print(f'"python " does not exists in {string}')
```

## Câu lệnh if lồng nhau

```python
# Nhập vào tổng số giờ làm và level của nhân viên. Nhân viên sẽ được khen thưởng nếu như tổng số giờ làm > 800 giờ. 

# Mức độ khen thưởng như sau:

# - Nếu là nhân viên level 1 thì thưởng 1000

# - Nếu là nhân viên level 2 thì thưởng 2000
total_hour = int(input("input hour :"))
total_level = int(input("input level: "))
if total_hour > 800 :
    if total_level == 1:
        print ("nhan vien duoc thuong 1000")
    if total_level ==2 :
        print("nhan vien duoc thuong 2000")
else:
    print("ban khong duoc nhan thuong ")

    
```

## Câu lệnh if…elif…else

```python
# Viết chương trình nhập vào số ngày trong tuần (2, 3, 4, 5, 6, 7, 8) và in ra từ tiếng Anh tương ứng với ngày đó.
```

```python
day = str(input("input a day on a week : "))
if day == "2":
    print("monday")
elif day == "3" :
    print("tuesday")
elif day == "4" :
    print("wednesday")
elif day == "5":
    print("thursday")
elif day == "6":
    print("friday")
elif day == "7":
    print("saturday")
elif day == "8":
    print("sunday")
else:
    print("wrong. Please input again ")
    
```
