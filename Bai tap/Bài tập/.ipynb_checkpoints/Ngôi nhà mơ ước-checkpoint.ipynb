```python
import turtle
pen = turtle.Turtle()
turtle.bgcolor("light blue")
pen.speed(5)
pen.pensize(10)

#ve dat
pen.penup()
pen.goto(-1000,-600)
pen.fillcolor("green")
pen.begin_fill()
pen.pendown()
pen.left(90)
pen.forward(500)
pen.right(90)
pen.forward(2000)
pen.right(90)
pen.forward(500)
pen.right(90)
pen.forward(2000)
pen.end_fill()
pen.left(90)




# ve nha
pen.penup()
pen.pencolor("black")
pen.goto(-200,-100)
pen.pendown()
pen.left(180)
pen.forward(300)
pen.fillcolor("orange")
pen.begin_fill()
pen.left(90)
pen.forward(60)
pen.right(120)
pen.forward(100)
pen.right(60)
pen.forward(400)
pen.right(60)
pen.forward(100)
pen.end_fill()
pen.right(120)
pen.forward(450)
pen.penup()
pen.goto(180,200)
pen.left(90)
pen.pendown()
pen.forward(300)
# ve cua so
pen.penup()
pen.fillcolor("yellow")
pen.begin_fill()
pen.goto(85,140)
pen.pendown()
pen.forward(60)
pen.left(90)
pen.forward(60)
pen.left(90)
pen.forward(60)
pen.left(90)
pen.forward(60)
pen.end_fill()

pen.penup()
pen.goto(-85,140)
pen.fillcolor("yellow")
pen.begin_fill()
pen.pendown()
pen.forward(60)
pen.left(90)
pen.forward(60)
pen.left(90)
pen.forward(60)
pen.left(90)
pen.forward(60)
pen.end_fill()
# ve cua 
pen.penup()
pen.goto(-40,-100)
pen.fillcolor("yellow")
pen.begin_fill()
pen.pendown()
pen.forward(100)
pen.right(90)
pen.forward(70)
pen.right(90)
pen.forward(100)
pen.end_fill()
# ve cay
pen.penup()
pen.goto(370,220)
pen.pendown()
pen.color("green")
pen.fillcolor("green")
pen.begin_fill()


pen.circle(100)
pen.end_fill()
pen.color("brown")
pen.penup()
pen.goto(470,110)
pen.pensize(30)
pen.pendown()
pen.forward(200)


# ve mat troi
pen.color("yellow")
pen.pensize(15)
pen.penup()
pen.goto(-500,300)
pen.pendown()
pen.fillcolor("red")
pen.begin_fill()
pen.circle(60)
pen.end_fill()





















turtle.done()
```
