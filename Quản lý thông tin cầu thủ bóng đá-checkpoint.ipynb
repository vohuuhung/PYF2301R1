```python
class Data:
    def __init__(self,season,name,position,year,nation,club,appearance,goal,assist):
        self.season = season
        self.name = name
        self.position =position
        self.year = year
        self.nation = nation
        self.club = club
        self.appearance = appearance
        self.goal = goal
        self.assist = assist
from tkinter import *
from tkinter.messagebox import showinfo
from tkinter import ttk
# Tạo cửa sổ chính của ứng dụng
window = Tk()
window.title("Quản lý thông tin cầu thủ")
window.geometry("1200x800")
list_person =[]
def add_person():
    # Hàm này sẽ được gọi khi nút "Thêm" được nhấn
    # Lấy dữ liệu từ các ô nhập liệu
    season_value = season.get()
    name_value = name.get()
    position_value=position.get()
    year_value = year.get()
    nation_value = nation.get()
    club_value = club.get()
    appearance_value = appearance.get()
    goal_value = goal.get()
    assist_value = assist.get()
    # Thêm dữ liệu vào bảng
    treeview.insert('', 'end', values=(season_value,name_value,position_value,year_value,nation_value,club_value,appearance_value,goal_value,assist_value))

    # Xóa dữ liệu trong các ô nhập liệu
    season.set("")
    name.delete(0, 'end')
    position.set("")
    year.set("")
    nation.delete(0, 'end')
    club.delete(0, 'end')
    appearance.delete(0, 'end')
    goal.delete(0, 'end')
    assist.delete(0, 'end')
    person = Data(season_value,name_value,position_value,year_value,nation_value,club_value,appearance_value,goal_value,assist_value)
    list_person.append(person)

# Tạo hàm tìm kiếm
def search_person():
    search_name_value = search_name.get() 
    search_season_value = search_season.get()
    for data in list_person:
        if search_name_value in data.name and search_season_value in data.season :
            showinfo("Search result", f"{data.season}\n{data.name} - {data.position} - {data.year}\n{data.nation} - {data.club}\nSố lần ra sân:{data.appearance}\nBàn thắng:{data.goal} - Kiến tạo:{data.assist}")
            break
            
# Tạo hàm sửa
def edit_person():
    season_value = season.get()
    name_value = name.get()
    position_value=position.get()
    year_value = year.get()
    nation_value = nation.get()
    club_value = club.get()
    appearance_value = appearance.get()
    goal_value = goal.get()
    assist_value = assist.get()
    selected_item = treeview.focus()
    treeview.item(selected_item,values =(season_value,name_value,position_value,year_value,nation_value,club_value,appearance_value,goal_value,assist_value))
# Tạo hàm xóa
def delete_person():
    selected_item = treeview.focus()
    treeview.delete(selected_item)
# Tạo treeview
treeview = ttk.Treeview(window, columns=("Mùa giải","Tên cầu thủ","Vị trí","Năm sinh","Quốc tịch","Đội bóng","Số lần ra sân","Bàn thắng","Kiến tạo"))
treeview.heading("Mùa giải", text="Season")
treeview.heading("Tên cầu thủ", text="Player name")
treeview.heading("Vị trí", text="Position")
treeview.heading("Năm sinh", text="Year of Birth")
treeview.heading("Quốc tịch", text="Nation")
treeview.heading("Đội bóng", text="Club")
treeview.heading("Số lần ra sân", text="Appearance")
treeview.heading("Bàn thắng", text="Goal")
treeview.heading("Kiến tạo", text="Assist")
treeview.place(x=50,y=100)
# Chỉnh độ rộng của cột
for column in treeview['columns']:
    treeview.column(column,width=100)


# Chỉnh sửa cột thông tin    
frame = Frame(window)
frame.place(x=150,y=350)
user_info_frame = LabelFrame(frame,text="Football player information!",bg="palegreen")
user_info_frame.grid(row=0,column=0)
# label
season_lbl = Label(user_info_frame,text="Season :",font=("arial",11),bg="palegreen")
season_lbl.grid(row=0,column=0,padx=20,pady=20)
name_lbl = Label(user_info_frame, text="Player name :",font=("arial",11),bg="palegreen")
name_lbl.grid(row=2,column=0,padx=20,pady=20)
position_lbl = Label(user_info_frame,text="Position :",font=("arial",11),bg="palegreen")
position_lbl.grid(row=4,column=0,padx=20,pady=20)
year_lbl = Label(user_info_frame,text="Year of birth :",font=("arial",11),bg="palegreen")
year_lbl.grid(row=6,column=0,padx=20,pady=20)
nation_lbl = Label(user_info_frame, text="Nation :",font=("arial",11),bg="palegreen")
nation_lbl.grid(row=8,column=0,padx=20,pady=20)
club_lbl = Label(user_info_frame, text="Club : ",font=("arial",11),bg="palegreen")
club_lbl.grid(row=10,column=0,padx=20,pady=20)
appearance_lbl = Label(user_info_frame, text="Appearance: ",font=("arial",11),bg="palegreen")
appearance_lbl.grid(row=2,column=4,padx=20,pady=20)
goal_lbl = Label(user_info_frame,text="Goal :",font=("arial",11),bg="palegreen")
goal_lbl.grid(row=4,column=4,padx=20,pady=20)
assist_lbl = Label(user_info_frame,text="Assist : ",font=("arial",11),bg="palegreen")
assist_lbl.grid(row=6,column=4,padx=20,pady=20)
# # Entry và combobox

season = ttk.Combobox(user_info_frame,values=[f"{i}-{i+1}" for i in range(1950, 2023)])
season.grid(row=0,column=2,padx=20,pady=20)
name = Entry(user_info_frame)
name.grid(row=2,column=2,padx=20,pady=20)
position = ttk.Combobox(user_info_frame,values =["FW","MF","DF","GK"])
position.grid(row=4,column=2,padx=20,pady=20)
year = ttk.Combobox(user_info_frame,values=list(range(1950,2023)))
year.grid(row=6,column=2,padx=20,pady=20)
nation = Entry(user_info_frame)
nation.grid(row=8,column=2,padx=20,pady=20)
club = Entry(user_info_frame)
club.grid(row=10,column=2,padx=20,pady=20)
appearance = Entry(user_info_frame)
appearance.grid(row=2,column=6,padx=20,pady=20)
goal = Entry(user_info_frame)
goal.grid(row=4,column=6,padx=20,pady=20)
assist = Entry(user_info_frame)
assist.grid(row=6,column=6,padx=20,pady=20)
# Tạo Label tên đề tài:
concert_lbl = Label(window,text="FOOTBALL PLAYER INFORMATION MANAGEMENT SYSTEM",bg="palegreen",font=("Arial","16"),width=80)
concert_lbl.place(x=150,y=50)
# Tạo nút tìm kiếm
frame1 = Frame(window)
frame1.place(x=800,y=350)
user_info_frame1 = LabelFrame(frame1,text="Input name and season of football player !",bg ="palegreen")
user_info_frame1.grid(row=0,column=0)

search_name = Entry(user_info_frame1)
search_name_btn = Button(user_info_frame1,text="Search",font=("arial",12,"bold"),bg="darkslategray",fg="white",command=search_person)
search_name.grid(row=0,column=3,padx=50,pady=20)

search_season = ttk.Combobox(user_info_frame1,values=[f"{i}-{i+1}" for i in range(1950, 2023)])
search_season.grid(row=1,column=3,padx=50,pady=20)
search_name_btn.grid(row=2,column=3,padx=50,pady=20)

# Tạo frame chứa Thêm xóa ,edit
frame2 = Frame(window)
frame2.place(x=870,y=570)
user_info_frame2 = LabelFrame(frame2,bg ="palegreen")
user_info_frame2.grid(row=0,column=0)
# Tạo nút "Thêm"
button_add = Button(user_info_frame2, text="Add",font=("arial",12,"bold"),bg="darkslategray",fg="white", command=add_person)
button_add.grid(row=0,column=0,padx=20,pady=10)
# Tạo nút xóa
delete_btn = Button(user_info_frame2,text = "Delete",font=("arial",12,"bold"),bg="darkslategray",fg="white",command=delete_person)
delete_btn.grid(row=2,column=0,padx=20,pady=10)
# Tạo nút edit
edit_btn = Button(user_info_frame2,text = "Edit",font=("arial",12,"bold"),bg="darkslategray",fg="white",command = edit_person)
edit_btn.grid(row=4,column=0,padx=20,pady=10)
window.mainloop()
```
